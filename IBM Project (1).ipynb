# Algorithms 

# Spliting up the data

```python
df.columns
```

```python
df = df[['male', 'age', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose', 'TenYearCHD']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['male', 'age', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose']].values
Y = df[['TenYearCHD']].values
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rmc = RandomForestClassifier()
rmc
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
rmc.fit(X_train, Y_train)
```

```python
rf_pred = rmc.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(rf_pred,Y_test)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, Y_train)
s_pred = svm.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(s_pred,Y_test)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
lor_pred = lr.predict(X_test) 
from sklearn.metrics import accuracy_score
accuracy_score(lor_pred,Y_test)
```

# Accuracy scores of all algorithms

```python
print("Accuracy scores of each model")
print("Random Forest : ", accuracy_score(rf_pred,Y_test))
print("SVM : ", accuracy_score(s_pred,Y_test))
print("Logistic regression : ", accuracy_score(lor_pred,Y_test))
print()
print("All algorithms have same accuracy ")
```
